```python
import numpy as np
from copulas import plot_random_variable, plot_amh, plot_contour, plot_contour2d
```

```python
# theta is in [-1, 1]
theta = 0.9
```

```python
def c(u, v):
    return (u * v) / (1 - (theta * (1 - u) * (1 - v)))
```

```python
plot_contour(c, 'Ali-Mikhail-Haq')
```

```python
plot_contour2d(c, 'Ali-Mikhail-Haq')
```
