```python
import numpy as np
from copulas import plot_random_variable, plot_fgm, plot_contour, plot_contour2d, plot_surface
```

```python
# theta is in [-1, 1]
theta = 0.5
```

```python
def c(u, v):
    return u * v + theta * v * (1 - v) * u * (1 - u)
```

```python
plot_contour(c, 'Farlie-Gumbel-Morgenstern')
```

```python
plot_surface(c, 'Farlie-Gumbel-Morgenstern')
```

```python
plot_contour2d(c, 'Farlie-Gumbel-Morgenstern')
```

```python
u = np.random.default_rng().uniform(0, 1, 5000)
t = np.random.default_rng().uniform(0, 1, 5000)
```

```python
plot_random_variable(u, t)
```

```python
plot_fgm([-1, 0, 1], u, t, c)
```
